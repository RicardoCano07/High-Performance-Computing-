```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    a=0.3
    v=2.0/3.0
    w=2.0
    b=1.0/1.3
    c=1.0/30.0
    return a*np.sin(np.cos(v-w*x))+b*x-c*x**2
```

```python
x = np.linspace(0, 25, 100)
plt.plot(x,func(x))
plt.grid()
plt.title('Plot of $f(x)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```
